```python
import matplotlib.pyplot as plt
import numpy as np


def addTwo(i):
    return i**3 - 3*i + 2
   
applyall = np.vectorize(addTwo)

grange = 3
acccur = 0.1

xs = np.arange(-grange, grange, acccur)
ys = applyall(np.arange(-grange, grange, acccur))



plt.plot(xs, ys)
plt.ylabel('some numbers')
plt.show()
```
